```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
import pandas as pd
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
```

```python
data = pd.read_csv('data-hasil-penertiban-minuman-keras-satpol-pp-jakarta-utara-periode-juni-2017.csv')
data
```

```python
fitur = data.iloc[:, :-1]
label = data.iloc[:, -1]
```

```python
fitur
```

```python
label
```

```python
le = preprocessing.LabelEncoder()
fitur = fitur.apply(le.fit_transform)
```

```python
le
```

```python
fitur
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(fitur, label)
```

```python
prediksi = clf.predict(fitur)
```

```python
prediksi
```

```python
from sklearn.metrics import accuracy_score, classification_report

akurasi = accuracy_score(label, prediksi)
laporan_klasifikasi = classification_report(label, prediksi)
```

```python
akurasi
```

```python
laporan_klasifikasi
```

```python
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=fitur.columns.values,
                           class_names=label.unique(),
                           filled=True, rounded=True,
                           special_characters=True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
graph
```
